## Libraries

```python
from sklearn.model_selection import KFold, cross_val_score, RandomizedSearchCV, cross_validate
from xgboost import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## Model Selection

### K Fold
Performing k fold validation for selecting model

```python
n_folds = 5
def rmse_accu_cv(model):
    #kf = KFold(n_splits=n_folds, shuffle=True, random_state=42).get_n_splits(inputs.values)
    rmse = np.sqrt(-cross_val_score(model, inputs, target, scoring='neg_mean_squared_error', cv=5))
    return rmse
```

```python
#models
xgb = XGBRegressor()
dt = DecisionTreeRegressor()
lgbm = LGBMRegressor()
rm = RandomForestRegressor()
gb = GradientBoostingRegressor()
lr = LinearRegression()
models = [xgb, dt, lgbm, rm, gb, lr]
models_name = {0:'XGB', 1:'DT', 2:'LGBM', 3:'RM', 4:'GB', 5:'LR'}
```

```python
for i in range(len(models)):
    rmse = rmse_accu_cv(models[i])
    print(f'{models_name[i]}: Min rmse = {min(rmse)}, Max rmse = {max(rmse)}, Mean rmse = {rmse.mean()}')
```

## Randomized Search CV with Hypertuning
Performing hyperparameter tuning to selected algos

```python
final_models = [xgb, lgbm, rm, gb]
final_models_name = {0:'XGB', 1:'LGBM', 2:'RM', 3:'GB'}
```

```python
def cross_validate_hypertuning(model, params, no_iter, inputs, target):
    rdmsearch = RandomizedSearchCV(model, param_distributions=params, n_jobs=-1,
                                   scoring='neg_root_mean_squared_error', n_iter=no_iter, cv=5)
    rdmsearch.fit(inputs, target)
    ht_params = rdmsearch.best_params_
    ht_score = rdmsearch.best_score_
    return ht_params, ht_score
```

```python
xgb_params ={
    'n_estimators': [100, 200, 300, 500, 1000],
    'learning_rate': [0.05],
    'booster': ['gbtree', 'gblinear', 'dart'],
    'max_depth': [3, 5, 10]
}

lgbm_params ={
    'n_estimators': [50, 100, 200, 300, 500, 1000],
    'learning_rate': [0.05],
    'boosting_type': ['gbdt', 'dart'],
    'num_leaves': [31, 50, 100],
    'max_depth': [3, 5, 10]
}

rm_params ={
    'n_estimators': [100, 200, 300, 500, 1000],
    'max_features': ['auto', 'sqrt'],
    'min_samples_split': [2, 5, 10],
    'bootstrap': [True, False],
    'max_depth': [3, 5, 10]
}

gb_params ={
    'n_estimators': [50, 100, 200, 300, 500, 1000],
    'max_depth': [3, 5, 10]
}
final_models_params = [xgb_params, lgbm_params, rm_params, gb_params]
```

```python
for i in range(len(final_models)):
    parameter, score = cross_validate_hypertuning(final_models[i], 
                                                  final_models_params[i], 50, inputs, target)
    print(f'{final_models_name[i]} parameter: {parameter}')
    print(f'{final_models_name[i]} Score: {score}')
```

## Stacking Models

I've selected 3 best models and stack them to make predictions

```python
def train_and_predict(inputs, target, prediction_data):
    xgb = XGBRegressor(n_estimators=500, max_depth=3, learning_rate=0.05, booster='gbtree')
    lgbm = LGBMRegressor(num_leaves=31, n_estimators=300, max_depth=3, 
                         learning_rate=0.05, boosting_type='gbdt')
    gb = GradientBoostingRegressor(n_estimators=200, max_depth=3)
    
    #train models
    xgb.fit(inputs, target)
    xgb_predict = xgb.predict(prediction_data)
    
    lgbm.fit(inputs, target)
    lgbm_predict = lgbm.predict(prediction_data)
    
    gb.fit(inputs, target)
    gb_predict = gb.predict(prediction_data)
    
    with open("House_price_predict_models.pkl", "wb") as f:
            pickle.dump([xgb, lgbm, gb], f)
    
    final_predictions = (np.expm1(xgb_predict) + np.expm1(lgbm_predict) + np.expm1(gb_predict))/3
    return final_predictions
```

```python
predictions = pd.DataFrame(train_and_predict(inputs, target, test_data), columns=['Sales'])
```

```python
#function for removing nested lists of predictions
import collections
def flatten(x):
    result = []
    for el in x:
        if isinstance(x, collections.Iterable) and not isinstance(el, float):
            result.extend(flatten(el))
        else:
            result.append(el)
    return result
```

```python
#Scaling predictions from between 0,1 to original values
predictions = scale_output.inverse_transform(predictions)
```
